```python
# drop the rows where destination is null
airlinedata.dropna(subset=['Destination'],inplace = True)

# drop the other destinations
airlinedata = airlinedata[airlinedata['Destination']!= 4]

# dropping the rows where airline is null
airlinedata.dropna(subset=['Airline'],inplace =  True)
```

```python
# airport choice data processing
# create dummy variables for airport choice
airport = airlinedata[['Airport','Airline','Age','Gender','Destination','DepartureHr',
                       'DepartureTime','AccessTime','ProvinceResidence','ModeTransport',
                       'NoTransport','FlyingCompanion','Nationality']]
```

```python
# checking the airport choice based on the mode of transport
print(airport[['Airport','ModeTransport','Age']].groupby(['ModeTransport','Airport']).count())
```

```python
# categorizing the mode of transport
airport['TransportCat'] = airport['ModeTransport'].apply(lambda x: 1 if x == 1 or x==2 or x==3 
                                                         else (2 if x==4 or x==5 
                                                         else(3 if x==6 or x==7 or x==8 else 4)))

# creating dummy variables
df1 = pd.DataFrame()
df1[['KoreanAir','Asiana','KoreanLCC','Foreign']] = pd.get_dummies(airport['Airline'])
df1[['Inchoen','Gimpo']] = pd.get_dummies(airport['Airport'])
df1[['Male','Female']] = pd.get_dummies(airport['Gender'])
df1[['Southeast AsiaD','ChinaD','JapanD']] = pd.get_dummies(airport['Destination'])
df1[['PersonalRoad','PublicSlow','PublicFast','SpecialityOthers']] = pd.get_dummies(airport['TransportCat'])
df1[['Korea','ChinaN','JapanN','SoutheastAsiaN','OtherN']] = pd.get_dummies(airport['Nationality'])


# merging the created dummy variables with the airport data
airport = pd.concat([airport,df1],axis = 1)

```

```python
# checking the sample size to fill the missing values for departure hour
print(airport[['Destination','DepartureTime','DepartureHr']].groupby(['Destination','DepartureTime']).count())
```

```python
# finding the median departure hour
print(airport[['Destination','DepartureTime','DepartureHr']].groupby(['Destination','DepartureTime']).agg(lambda x: x.mode()[0]).reset_index())
```

```python

# filling the missimng values for departure hour
depart = airport[['Destination','DepartureHr','DepartureTime']]
depart['DepartureHr'] = pd.to_numeric(depart['DepartureHr'], errors='coerce')
DepHrProxy = depart[['Destination','DepartureTime','DepartureHr']].groupby(['Destination','DepartureTime']).agg(lambda x: x.mode()[0]).reset_index()
depart = depart.merge(DepHrProxy, on = ['Destination','DepartureTime'], how = 'left')
depart['DepartureHrFilled'] = depart['DepartureHr_x'].fillna(depart['DepartureHr_y'])

```

```python
# checking the sample size to fill the missing values for access time
print(airport[['AccessTime','ProvinceResidence']].groupby(['ProvinceResidence']).count())
```

```python
# finding the missing values to fill
print(airport[['AccessTime','ProvinceResidence']].groupby(['ProvinceResidence']).mean())
```

```python
# filling the missimng values for access time
access = airport[['AccessTime','ProvinceResidence','ModeTransport']]
AccessProxy = access.groupby(['ProvinceResidence','ModeTransport']).mean().reset_index()
AccessProxy['AccessTime'] = AccessProxy['AccessTime'].apply(lambda x: round(x,0))
access=access.merge(AccessProxy, on = ['ProvinceResidence','ModeTransport'],how='left')
access['AccessTimeFill'] = access['AccessTime_x'].fillna(access['AccessTime_y'])

# filling the missimng values for access time
time = access[['ProvinceResidence','AccessTimeFill']]
proxy = time.groupby(['ProvinceResidence']).mean().reset_index()
proxy['AccessTimeFill'] = proxy['AccessTimeFill'].apply(lambda x: round(x,0))
time = time.merge(proxy, on =['ProvinceResidence'],how = 'left')
time['AccessTimeFilled'] = time['AccessTimeFill_x'].fillna(time['AccessTimeFill_y'])


# adding the new columns without missing values to the airport data

airport.reset_index(drop = True, inplace = True)
depart.reset_index(drop = True, inplace = True)
time.reset_index(drop = True, inplace = True)
airport['DepartureHourFilled'] = depart['DepartureHrFilled']
airport['AccessTimeFilled'] = time['AccessTimeFilled']
```

```python
# exporting the processed files as csv
airport.to_csv(r"C:\Users\nisha\OneDrive\Desktop\Final Project\Final Data\airport_processed.csv", index = False)
```

```python
# airline and airport choice models
import pandas as pd
import numpy as np
import statsmodels.api as sm
from sklearn.model_selection  import train_test_split
from sklearn import metrics
from sklearn import tree
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC 
from sklearn.neural_network import MLPClassifier 
```

```python
airport = pd.read_csv(r"C:\Users\nisha\OneDrive\Desktop\Final Project\Final Data\airport_processed.csv")
```

```python
# remove unused columns from the airline data
airport.drop(columns = ['Airport','Gender','Destination','DepartureTime','ProvinceResidence',
                        'ModeTransport','Nationality','TransportCat','DepartureHr','AccessTime',
                       'Airline'], inplace = True)
pd.set_option('display.max_columns', 50)
```

```python
# airport choice model
airport['intercept'] = 1.0
y1 = airport['Inchoen']
x1 = airport[['intercept', 'Age', 'NoTransport', 'FlyingCompanion', 'Male', 'Southeast AsiaD', 'ChinaD', 'JapanD',
       'Korea', 'ChinaN', 'JapanN', 'SoutheastAsiaN',
       'DepartureHourFilled', 'AccessTimeFilled','KoreanAir','Asiana','KoreanLCC']]
xtrain1, xtest1, ytrain1, ytest1 = train_test_split(x1, y1, test_size = 0.20, random_state = 109)
```

```python
# logistic regression
logit1 = sm.Logit(ytrain1, xtrain1).fit()
print(logit1.summary())
```

```python
result1 = xtest1.copy()
pred1 = logit1.predict(xtest1)
result1.loc[:,'prediction']=0
result1.loc[pred1 > 0.8,'prediction']=1
print(pd.crosstab(ytest1, result1['prediction'],rownames =['Actual'],colnames=['Predicted']))
print("Accuracy:",metrics.accuracy_score(ytest1, result1['prediction']))
print("Precision:",metrics.precision_score(ytest1, result1['prediction']))
print("Recall:",metrics.recall_score(ytest1, result1['prediction']))
        
```

```python
# decision tree
y1 = airport['Inchoen']
x1 = airport[['Age', 'NoTransport', 'FlyingCompanion', 'Male', 'Southeast AsiaD', 'ChinaD', 'JapanD',
       'Korea', 'ChinaN', 'JapanN', 'SoutheastAsiaN',
       'DepartureHourFilled', 'AccessTimeFilled','KoreanAir','Asiana','KoreanLCC']]
xtrain1, xtest1, ytrain1, ytest1 = train_test_split(x1, y1, test_size = 0.20, random_state = 109)

clf1 = tree.DecisionTreeClassifier(class_weight=None, criterion='gini', max_depth=20, 
                                  max_features=None, max_leaf_nodes=None, min_samples_leaf=8,
                                  min_samples_split=4, min_weight_fraction_leaf=0.0, 
                                  random_state=100, splitter='best')
tree1 = clf1.fit(xtrain1, ytrain1)
ypred1 = tree1.predict(xtest1) 
print(pd.crosstab(ytest1, ypred1,rownames =['Actual'],colnames=['Predicted']))
print("Accuracy:",metrics.accuracy_score(ytest1, ypred1))
print("Precision:",metrics.precision_score(ytest1, ypred1))
print("Recall:",metrics.recall_score(ytest1, ypred1))
```

```python
# normalizing the variables
# normalizing the variables
scaler = StandardScaler()  
scaler.fit(xtrain1)
xtrain1 = scaler.transform(xtrain1)
xtest1 = scaler.transform(xtest1)
```

```python
# support vector machine
svclassifier = SVC(kernel='rbf')   
svclassifier.fit(xtrain1, ytrain1)  
ypred = svclassifier.predict(xtest1) 
print(pd.crosstab(ytest1, ypred,rownames =['Actual'],colnames=['Predicted']))
print("Accuracy:",metrics.accuracy_score(ytest1, ypred))
print("Precision:",metrics.precision_score(ytest1, ypred))
print("Recall:",metrics.recall_score(ytest1, ypred))
```

```python
# neural network
mlp = MLPClassifier(hidden_layer_sizes=(30,30,30,30), max_iter=1000)  
mlp.fit(xtrain1, ytrain1)  
ypred = mlp.predict(xtest1) 
print(pd.crosstab(ytest1, ypred,rownames =['Actual'],colnames=['Predicted']))
print("Accuracy:",metrics.accuracy_score(ytest1, ypred))
print("Precision:",metrics.precision_score(ytest1, ypred))
print("Recall:",metrics.recall_score(ytest1, ypred))
```
